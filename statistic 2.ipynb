QUESTION 6) The population mean salary is RS 50,000 , with  σ RS =5000
            if we take a sample of 100 employees, what is the standard errors
            of the mean (SEM)

ANSWER:     Population standard deviation (σ) = 5000

            Sample size n=100

             SEM = σ / √n

             = 5000 / 10 = 500
